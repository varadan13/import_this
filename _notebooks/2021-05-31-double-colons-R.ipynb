**What are the double colons (::) in R?**

:: helps to access the exact function from that specific package.

When dplyr or readr or any package is loaded R sometimes gives out a message as follows.   
> The following objects are masked from ‘package:base’:   

so to use a specific object from a specific package use   

package_name::object_name

```r
(.packages()) # a neat trick to view 
              # all the loaded packages 
              # in a session.
```

- Notice from the above output ggthemes is the last package added to the current session and to load any object R looks at ggthemes first and then searches in forcats and so on. 

- Also notice that the above output are ordered (the last added package is displayed first and then the next last and so on)
