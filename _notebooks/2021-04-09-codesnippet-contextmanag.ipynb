```python
def a_simple_context_func(file_address,message,shld_write=True):
  file=open(file_address,"wt")
  if shld_write==True:
    file.__enter__()
    file.write(message)
    file.__exit__()
```

```python
Context=a_simple_context_func("/content/sample_data/README.md","a simple message to write")
```
